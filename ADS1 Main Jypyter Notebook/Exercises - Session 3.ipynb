In this exercise, we will experiment with the time complexity of adding elements to a HashMap vs a TreeMap. Python includes a HashMap in its standard library:

    > hashmap = dict()
and you can assign a (key, value) pair by writing:

    > hashmap[key] = value
    
Python does not include a TreeMap in its standard library, but you can get one from the module sortedcontainers: 

    > from sortedcontainers import SortedDict
    > treemap = SortedDict()
    
__Pairs of keys and values can be assigned as above.__

__Now for the exercise:__

__a)__	How long time should it take, in Big-Oh notation, to insert N elements into an initially empty HashMap and TreeMap, respectively.

__Answer:__
    
    HashMap should take O(1) time (constant) to insert into an initially empty HashMap, due to us inserting on a specific key location. As the HashMap fills, the time will decrease. Worst case it will take O(N) time in a almost full HashMap.

 _

    TreeMap takes O(N*log(N)) time (logarithmic) to insert N elements.
    

b)	Write a python script that inserts 10000 elements into an initially empty HashMap and TreeMap, respectively. Measure the time it takes. Which one is faster?



```python
from time import time
from sortedcontainers import SortedDict
import matplotlib.pyplot as plt


# Define a function to insert and time the HashMap
def insertIntoHashMap(elements, hashmap):
    start_time = time()

    for element in elements:
        hashmap[element] = time() - start_time

    end_time = time()
    return end_time - start_time


# Define a function to insert and time the TreeMap
def insertIntoTreeMap(elements, treeMap):
    start_time = time()

    for element in elements:
        treeMap[element] = time() - start_time
        
    end_time = time()
    return end_time - start_time


# Lets test the two functions:
numberOfElements = 1000000

elements = []
for i in range(numberOfElements):
    elements.insert(i, i)

hashMap = dict()
treeMap = SortedDict()
timeHashMap = insertIntoHashMap(elements, hashMap)
timeTreeMap = insertIntoTreeMap(elements, treeMap)

print("Hashmap time for inserting {} elements into empty Hashmap (dict) is       : {} seconds".format(numberOfElements, timeHashMap))
print("TreeMap time for inserting {} elements into empty TreeMap (SortedDict) is : {} seconds".format(numberOfElements, timeTreeMap))
```
